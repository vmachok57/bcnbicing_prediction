```python
import dask
import dask.dataframe as dd
import dask.array as da
```

```python
#Importamos todos los archivos .csv con que vamos a trabajar
df = dd.read_csv("ProyectoF/data/data_fit/*.csv", dtype={'last_reported': 'float64','station_id':'float64','num_docks_available':'float64','is_returning':'float64','ttl':'float64'})
```

```python
df.head()
```

```python
#Transformamos en un formato de hora legible
df["last_reported"] = dd.to_datetime(df["last_reported"], unit='s')
```

```python
df.head()
```

```python
#Descomponemos el Datatime (last_reported)
df["year"] = df["last_reported"].dt.year
df["month"] = df["last_reported"].dt.month
df["day"] = df["last_reported"].dt.day
df["hour"] = df["last_reported"].dt.hour
```

```python
df.head()
```

```python
df_S1 = df[(df["station_id"] == 1) & (df["year"] == 2021) & (df["month"] == 4) & (df["day"] == 5)].compute()
df_S1.head()
```

```python
# Verificar valores nulos en las columnas de agrupación
null_counts = df[['station_id', 'year', 'month', 'day', 'hour']].isnull().sum().compute()
print(null_counts)
```

```python
result = df.groupby(['station_id', 'year', 'month', 'day', 'hour'])['num_docks_available'].mean().reset_index()
```

```python
# Si `station_id`,'year','month','day','hour' es de tipo diferente, puedes convertirlo explícitamente
result[['station_id','year','month','day','hour']] = result[['station_id','year','month','day','hour']].astype(int)
```

```python
import pandas as pd
# Asegurar que 'result' es un DataFrame de Pandas
if not isinstance(result, pd.DataFrame):
    result = result.compute()  # Convertir de Dask a Pandas si es necesario

# Filtrar los datos
filtered_df_s1 = result[
    (result['station_id'] == 1) &
    (result['year'] == 2021) &
    (result['month'] == 1) &
    (result['day'] == 1)
]
# Mostrar las primeras 24 filas
filtered_df_s1 = filtered_df_s1.sort_values(by="hour", ascending=True)
#print(filtered_df_s1.head(24))
filtered_df_s1.head(24)
```

```python
# Unir los DataFrames usando 'station_id' como clave
merged_df = pd.merge(result, df_info_station, on="station_id", how="left")

# Mostrar las primeras filas del resultado
merged_df.head()
```

```python
# Agregar la columna "percentage_docks_available"
merged_df["percentage_docks_available"] = merged_df["num_docks_available"] / merged_df["capacity"]

# Opcional: llenar valores NaN si hay estaciones sin capacidad registrada
merged_df["percentage_docks_available"].fillna(0, inplace=True)
```

```python
# Ordenar los datos correctamente antes de agrupar
df_prediction_task = df_prediction_task.sort_values(
    by=['station_id', 'year', 'month', 'day', 'hour']
)
```

```python
# Crear las columnas desplazadas sin perder conexión entre días
for i in range(1, 5):
    df_prediction_task[f'disponibilidad_porcentage_{i}h_antes'] = (
        df_prediction_task.groupby('station_id')['percentage_docks_available']
        .shift(i)  # Desplazamiento de i horas
        .fillna(0)  # Reemplaza NaN con 0
    )
```
